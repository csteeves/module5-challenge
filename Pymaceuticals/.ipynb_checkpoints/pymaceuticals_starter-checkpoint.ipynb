```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


```

```python
# viewing the data sets
mouse_metadata.head()
```

```python
# viewing the data sets
study_results.head()
```

```python
# Combine the data into a single dataset
mouse_study = pd.merge(mouse_metadata, study_results,
                                 how='outer', on='Mouse ID')

# Display the data table for preview
mouse_study.head(15)
```

```python
# Checking the number of mice.
len(mouse_study["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = mouse_study.loc[mouse_study.duplicated(subset=["Mouse ID","Timepoint",]),"Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_data = mouse_study.loc[mouse_study["Mouse ID"] == "g989",:]
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mice = mouse_study.loc[mouse_study["Mouse ID"]!= "g989",:]
clean_mice.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_mice["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
regimen_mean = clean_mice.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
regimen_median= clean_mice.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
regimen_variance = clean_mice.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
regimen_sd = clean_mice.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
regimen_sem = clean_mice.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# Assemble the resulting series into a single summary DataFrame.
statistic_sum = pd.DataFrame({
    "Mean Tumor Volume":regimen_mean,
    "Median Tumor Volume": regimen_median,
    "Tumor Volume Variance": regimen_variance,
    "Tumor Volume Std. Dev.": regimen_sd,
    "Tumor Volume Std. Err.": regimen_sem
})
statistic_sum
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
statistic_agg = clean_mice.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean","median","var","std","sem"])
statistic_agg
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
mice_regimen = clean_mice.groupby(["Drug Regimen"]).count().sort_values("Timepoint",ascending=False)
mice_regimen = pd.DataFrame(mice_regimen["Timepoint"])
figure1 = mice_regimen.plot(kind="bar",xlabel="Drug Regimen",ylabel="Number of Mice Tested")
figure1
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
mice_regimen = clean_mice["Drug Regimen"].value_counts()
x_axis = mice_regimen.index.values
y_axis = mice_regimen.values
plt.bar(x_axis,y_axis,align="center",color="purple")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Mice Tested")
plt.xticks(rotation="vertical")
plt.tight_layout()
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
mice_sex = clean_mice["Sex"].value_counts()
mice_sex.plot(kind="pie",autopct="%1.1f%%")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(mice_sex,autopct="%1.1f%%",labels=mice_sex.index.values)
plt.ylabel("Sex")
plt.show()
```
