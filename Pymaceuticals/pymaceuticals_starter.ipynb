```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


```

```python
# viewing the data sets
mouse_metadata.head()
```

```python
# viewing the data sets
study_results.head()
```

```python
# Combine the data into a single dataset
mouse_study = pd.merge(mouse_metadata, study_results,
                                 how='outer', on='Mouse ID')

# Display the data table for preview
mouse_study.head(15)
```

```python
# Checking the number of mice.
len(mouse_study["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = mouse_study.loc[mouse_study.duplicated(subset=["Mouse ID","Timepoint",]),"Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_data = mouse_study.loc[mouse_study["Mouse ID"] == "g989",:]
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mice = mouse_study.loc[mouse_study["Mouse ID"]!= "g989",:]
clean_mice.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_mice["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
regimen_mean = clean_mice.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
regimen_median= clean_mice.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
regimen_variance = clean_mice.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
regimen_sd = clean_mice.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
regimen_sem = clean_mice.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# Assemble the resulting series into a single summary DataFrame.
statistic_sum = pd.DataFrame({
    "Mean Tumor Volume":regimen_mean,
    "Median Tumor Volume": regimen_median,
    "Tumor Volume Variance": regimen_variance,
    "Tumor Volume Std. Dev.": regimen_sd,
    "Tumor Volume Std. Err.": regimen_sem
})
statistic_sum
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
statistic_agg = clean_mice.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean","median","var","std","sem"])
statistic_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_mice.groupby(["Mouse ID"])["Timepoint"].max()
last_timepoint = last_timepoint.reset_index()
# last_timepoint

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_last_timepoint = last_timepoint.merge(mouse_study,on = ["Mouse ID","Timepoint"])
merge_last_timepoint.head()
```

```python
# Filtering the merged data set for the 4 treatment regimens 
merge_last_timepoint_regimen = merge_last_timepoint[(merge_last_timepoint["Drug Regimen"]=="Capomulin") | 
                                                        (merge_last_timepoint["Drug Regimen"]=="Ramicane") |
                                                        (merge_last_timepoint["Drug Regimen"]=="Infubinol") |
                                                        (merge_last_timepoint["Drug Regimen"]=="Ceftamin")
                                                       ]
# merge_last_timepoint_regimen.head()

# Creating new data frame to house only the Tumor Column and drug regimen
drug_tumor = merge_last_timepoint_regimen[["Drug Regimen","Tumor Volume (mm3)"]]

# drug_tumor.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    tumor_size = drug_tumor.loc[drug_tumor["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]

   # add subset
    tumor_vol.append(tumor_size)

# Determine outliers using upper and lower bounds
    quartiles = tumor_size.quantile([.25,.5,.75])
    lq= quartiles[.25]
    uq = quartiles[.75]
    iqr= uq-lq
    
    lower_b= lq- (1.5*iqr)
    upper_b= uq + (1.5*iqr)
    outlier = tumor_size.loc[(tumor_size < lower_b)  | (tumor_size > upper_b)]
    print(f" {treatment} has the potential outliers: {outlier}")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1= plt.subplots()
ax1.boxplot(tumor_vol,labels=treatments,flierprops= dict(markerfacecolor="red",markersize=5),)
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.set_xlabel("Drug Regimen")
plt.tight_layout()
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
average_tumor = capomulin.groupby(["Mouse ID"]).mean()
mouse_weight = average_tumor["Weight (g)"]
tumor_weight =average_tumor["Tumor Volume (mm3)"]
plt.scatter(mouse_weight,tumor_weight)
plt.xlabel("Weight (g)")
plt.ylabel("AverageTumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
(slope, interecept,rvalue,pvalue,stderr) = stats.linregress(average_tumor["Weight (g)"],average_tumor["Tumor Volume (mm3)"])
y_pred = slope*average_tumor["Weight (g)"]+interecept
```

```python
plt.scatter(mouse_weight,tumor_weight)
plt.plot(average_tumor["Weight (g)"],y_pred,color="red")
plt.xlabel("Weight (g)")
plt.ylabel("AverageTumor Volume (mm3)")

print(f'R-squared: {rvalue**2}')
plt.show()
```
